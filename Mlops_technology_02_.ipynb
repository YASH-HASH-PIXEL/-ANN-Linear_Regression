```python
# ml_pipeline.py – A CI/CD pipeline that retrains and deploys a model
import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 1. Load fresh data (simulates a new batch arriving in real time)
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42
)

# 2. Train a new model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# 3. CI gate: check accuracy
accuracy = accuracy_score(y_test, model.predict(X_test))
print(f"Model accuracy: {accuracy:.2f}")

# 4. Deploy if quality check passes
DEPLOY_THRESHOLD = 0.9
if accuracy >= DEPLOY_THRESHOLD:
    joblib.dump(model, "model.pkl")          # save for serving API
    print(" Model deployed successfully!")
else:
    print(" Accuracy below threshold. Pipeline aborted.")
```
